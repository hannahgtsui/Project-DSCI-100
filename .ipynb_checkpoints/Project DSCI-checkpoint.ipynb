## Topic： To what extent does economic vulnerability influence cryptocurrency adoption across countries?

Introduction： As traditional financial and employment systems face increasing strain in many parts of the world, cryptocurrency has emerged as an alternative financial tool. Rather than focusing on a single factor, this project examines whether broader forms of economic vulnerability—such as financial exclusion, labor market conditions, and income levels—are associated with higher levels of crypto adoption.

The goal is to understand whether crypto serves as a substitute for weak traditional systems, particularly in environments where access to banking or stable income is limited.

The data below is based on the World Bank, Chainalysis public information as references and will mainly scrutinize between 2022 to 2025 fluctuations of the data.

### Focus 1:  Do countries with higher inflation rates have higher crypto adoption?

Inflation reflects the degree of monetary instability within an economy, as rising prices reduce the purchasing power of a country’s currency. In such environments, individuals may be more likely to seek alternative stores of value or financial systems, such as cryptocurrency. To examine this relationship, we analyze cross-country data on inflation and crypto adoption.


```python
import pandas as pd
inflation= pd.read_csv("data/Inflation.csv",skiprows=3)
inflation["avg_inflation"] = inflation[["2022", "2023", "2024"]].mean(axis=1)
inflation = inflation[["Country Name", "avg_inflation"]].copy()
inflation["Country Name"] = inflation["Country Name"].str.strip()
inflation
```

```python
crypto= pd.read_csv("data/crypto_adoption_2024.csv")
crypto
```

```python
crypto = crypto.rename(columns={"country": "Country Name"})
```

```python
inflation_crypto = inflation.merge(crypto, on="Country Name", how="inner")
inflation_crypto
```

```python
inflation_crypto.sort_values(by="overall_rank_2024")
```

```python
import altair as alt
import pandas as pd

plot_data_filtered = inflation_crypto[[
    "Country Name",
    "avg_inflation",
    "overall_rank_2024",
    "region"
]].dropna()

plot_data_filtered = plot_data_filtered[plot_data_filtered["avg_inflation"] <= 30]

alt.Chart(plot_data_filtered).mark_circle(size=70).encode(
    x="avg_inflation:Q",
    y="overall_rank_2024:Q",
    color="region:N",
    tooltip=["Country Name", "avg_inflation", "overall_rank_2024"]
)
```

#### Conclusion of the above findings:
The relationship between inflation and cryptocurrency adoption does not appear strongly linear across countries. While some high-inflation countries show relatively higher adoption, the overall pattern remains dispersed, suggesting that inflation alone is not sufficient to explain differences in crypto adoption.


### Focus 3: Do combined economic vulnerability indicators better explain cryptocurrency adoption?

Since individual variables show limited explanatory power, we now consider a combined model of economic vulnerability. The following analysis examines whether multiple factors together better explain crypto adoption across countries.



```python
GDP= pd.read_csv("data/GDP.csv", skiprows=4)
GDP["avg_gdp"] = GDP[["2022", "2023", "2024"]].mean(axis=1)
GDP=GDP[["Country Name","2022", "2023", "2024","avg_gdp"]]
GDP
```

```python
bank=pd.read_csv("data/Bank access.csv",skiprows=3)
bank["bank_access"] = bank["2024"]
bank_clean = bank.dropna(subset=["bank_access"])
bank_clean = bank_clean[["Country Name", "bank_access"]]
```

```python
inflation_final = inflation[["Country Name", "avg_inflation"]].copy()
inflation_final["Country Name"] = inflation_final["Country Name"].str.strip()

crypto["Country Name"] = crypto["Country Name"].str.strip()

bank = bank[["Country Name", "bank_access"]].copy()
bank["Country Name"] = bank["Country Name"].str.strip()


GDP = GDP[["Country Name", "avg_gdp"]].copy()
GDP["Country Name"] = GDP["Country Name"].str.strip()


merged_all = inflation_final.merge(GDP, on="Country Name", how="inner") \
                            .merge(crypto, on="Country Name", how="inner") \
                            .merge(bank, on="Country Name", how="inner")

merged_all.head()
```

```python
model_data = merged_all[[
    "avg_inflation", 
    "avg_gdp", 
    "bank_access", 
    "overall_rank_2024"
]].dropna()

X = model_data[["avg_inflation", "avg_gdp", "bank_access"]]
y = model_data["overall_rank_2024"]
```

```python
import sys
!{sys.executable} -m pip install scikit-learn
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
model.coef_, model.intercept_
```

### Conclusion of the above findings:

The regression results indicate that inflation and financial inclusion play more significant roles in explaining cryptocurrency adoption than GDP alone. In particular, inflation shows a negative relationship with adoption rank (coefficient ≈ −0.26), suggesting that higher inflation is associated with increased crypto usage. Financial inclusion also exhibits a relatively strong effect (coefficient ≈ −0.57), while GDP contributes minimally (coefficient ≈ 0.00025). Overall, the combined model provides a more comprehensive explanation than any individual variable.

A point that is worth mentioning is that while inflation alone does not exhibit a strong linear relationship with crypto adoption, its effect becomes more apparent when controlling for other factors such as GDP and financial inclusion. This suggests that the relationship between inflation and crypto adoption is influenced by multiple interacting variables.


- #### Inflation:
Before combining inflation with other variables, we first examine its individual relationship with cryptocurrency adoption. Based on our initial hypothesis in Focus 1, we test whether inflation alone has a strong explanatory power. The results show that the relationship between inflation and crypto adoption is very weak (R² ≈ 0.004), indicating that inflation alone explains almost none of the variation.

To further evaluate the strength of the relationship, we examine the R² value of the model. 

```python
X1 = merged_all[["avg_inflation"]].dropna()
y1 = merged_all.loc[X1.index, "overall_rank_2024"]

model1 = LinearRegression().fit(X1, y1)
model1.score(X1, y1)
```

```python
model.score(X, y)
```

The result for inflation alone is extremely low (R² ≈ 0.004), indicating a very weak relationship. Even in the combined model, the R² remains relatively low (R² ≈ 0.063), suggesting that although the model improves, its overall explanatory power remains limited.
